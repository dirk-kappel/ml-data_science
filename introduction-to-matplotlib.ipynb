### Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
import matplotlib.pyplot as plt

# 1. Prepare the data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10,10))

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 5. Save and Show (you save the whole figure)
fig.savefig("matplotlib/sample-plot.png")
```

## Making figures with NumPy arrays
We want:
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplots

```python
import numpy as np
```

```python
# Create some data
# linspace - generate an array with a specific number of elements that are spaced evenly
x = np.linspace(0, 10, 100)
x
```

#### Line Plot

```python
# Plot the data and create line plot
fig, ax = plt.subplots()
ax.plot(x, x**2);  # plot is a line plot.
```

#### Scatter Plot

```python
# Use th same data to make a scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));  # scatter is the scatter plot
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x)); # Create a sin wave scatter plot
```

#### Bar Graph

```python
# make a plot from dictionary
nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Dan's Nut Butter Store",
       ylabel="Price ($)");
```

#### Horizontal Bar Graph

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values())); # Need to turn the values into a list for it to work.
```

#### Histogram

```python
# Make some data for histograms and plot it
x = np.random.randn(1000) # 1000 random samples that will fit in a normal distribution.
fig, ax = plt.subplots()
ax.hist(x);
```

#### Subplots
Two options

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5))
```

```python
# Plot to each different axis
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5))
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplot option 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10, 5))
# Plot to each different index
ax[0, 0].plot(x, x/2);
ax[1, 0].scatter(np.random.random(10), np.random.random(10));
ax[0, 1].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

### Which one should you use? (pyplot vs matplotlib OO method?)
* When plotting something quickly, okay to use pyplot method
* When plotting something more advanced, use the OO method

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
len(over_50)
```

```python
# Pyplot method
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target");
```

```python
# OO Method mixed with pyplot method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             ax=ax);
ax.set_xlim([45, 100]); # Set an axis limit between 45-100
```

```python
# OO Method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--");
```

## Subplot of chol, age, thalach

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10),
                               sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"]);

# Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
       ylabel="Cholesterol");

# Add a legend
ax0.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line
ax0.axhline(over_50["chol"].mean(),
          linestyle="--");

# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"]);

# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate",
       xlabel="Age",
       ylabel="Max Heart Rate");

# Add a legend
ax1.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line
ax1.axhline(over_50["thalach"].mean(),
          linestyle="--");

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
# Set the plot style
plt.style.use('seaborn-v0_8-whitegrid')

# OO Method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter") # This chages the color scheme

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--");
```

```python
# Customizing the ya nd x axis limitations

# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10),
                               sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"],
                      cmap="winter");

# Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
       ylabel="Cholesterol");

# Add a legend
ax0.legend(*scatter.legend_elements(), title="Target");
# Set the x limits
ax0.set_xlim([50, 80])
# Set the y limits
ax0.set_ylim([100, 500])
# Add a horizontal line
ax0.axhline(over_50["chol"].mean(),
          linestyle="--");

# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"],
                      cmap="winter");

# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate",
       xlabel="Age",
       ylabel="Max Heart Rate");

# Add a legend
ax1.legend(*scatter.legend_elements(), title="Target");
# Set the x limits
ax1.set_xlim([50, 80])
# Set the y limits
ax1.set_ylim([60, 200])
# Add a horizontal line
ax1.axhline(over_50["thalach"].mean(),
          linestyle="--");

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
